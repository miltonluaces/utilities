```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
min = -1000
max = 1000
X=np.array(range(min, max))/10
X
```

```python
f = lambda x : -np.e**(1.8*x)+7000
y = [f(x) for x in X]
plt.plot(y);
```

```python
f = lambda x : x/10
y = [f(x) for x in X]
plt.plot(y);
```

```python
a = 0.1
b = 1
c = 70
g = lambda x: a * math.e**(-((x-b)**2)/(2*c**2))
y = [g(x) for x in X]
plt.plot(y);                    
```
